```python
import glob
files = glob.glob(r'D:\Google雲端硬碟\NCKU\法經\法經報告\procedure\判決書\*\*.txt')
len(files)
```

```python
import re
filenames=[]
crime_nums= []
no=[]
tran_num = {u"零" : "0",u"一" : "1",u"二" : "2",u"三" : "3",u"四" : "4",u"五" : "5",u"六" : "6",u"七" : "7",u"八" : "8",
            u"九" : "9",'十':'','０':'0' ,'１':'1','２':'2','３':'3','４':'4','５':'5','６':'6','７':'7','８':'8','９':'9',
            u'壹': "1",u'貳': "2",u'參': "3",u'': "3",u'': "3",u'肆': "4",u'伍': "5",u'陸': "6",u'柒': "7",u'捌': "8",
            u'玖': "9",'○':'0'}

for i,file in enumerate(files):
    if i % 2700 == 0:
        print(i)
    with open(file,'r',encoding='utf-8-sig') as f:
        txt = f.read()
    part = txt.replace("\n","").replace(" ","")
    # 改成阿拉伯數字/
    for ele in tran_num:
        part = part.replace(ele ,tran_num[ele])
    # /改成阿拉伯數字
    input_text = part
    con1 ='[酒駕前科|公共危險案件|曾|已|有|復|屬|年間|年|再度|再犯|再|已是|已屬|係|為|是](\d+)[次|犯|度]'
    con2 = '(\d+)[次|犯|度][駕駛|酒駕|本案|本件]'
    regex =con1+'|'+con2
    pattern = re.compile(regex)
    match = re.findall(pattern, input_text)
    if match == []:
        no.append([i,file,part])
        continue
    max_ =0
    for sen in match: #[ ('','n'),... ]
        for i in sen:
            if i !='':
                t = int(i)
                if t > 50: #防抓錯
                    continue
                if t > max_:
                    max_ = t
    if max_ < 2:
        continue
    filename = file.split('\\')[-1]
    filenames.append(filename)
    crime_nums.append(max_)
        
```

```python
print(set(crime_nums))
len(no),len(crime_nums),len(filenames)
```

```python
import pandas as pd
df = pd.DataFrame(data={'index':filenames,'crime_nums': crime_nums})
path = 'D:/Google雲端硬碟/NCKU/法經/法經報告/procedure/csv/index_crime_nums.csv'
df.to_csv(path,encoding='utf-8-sig',index=False)
data = pd.read_csv('D:/Google雲端硬碟/NCKU/法經/法經報告/procedure/csv/data26885_6.csv',encoding='ansi')
data.shape
```
