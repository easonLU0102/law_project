## 總刑期
先定位出 1!yyy!mm!dd! 該行 \
定位出 '應執行刑' \d{4} \
無的話往下找出 1.1!刑法!185! 該行 \
定位出 'yymm' \
將 'yymm' 轉成 int(mm) \
note: 1!的行數為被告數 只有33件有多個被告排除討論

```python
case = cases[408]
print(case)
```

```python
def jail_int(num):
    year = num[:2]
    month = num[2:]
    y=int(year) *12
    m = int(month)
    return y+m
#總樣本為26928-33-10 = 26885
match_1 = {}
match_2 = {}
file_names =[]
jails = []
for i,case in enumerate(cases): #共26928件
    lines = case.split('\n')
    #/ 排除多位被告案, 使每個case只會有一個 '1!10' 共33件
    count = 0
    for line in lines:
        if line[:2] == '1!': #因終結案跟判決書日期可能不同
            count = count + 1
            #print(i,case)
    if count > 1: 
        continue
    # /
    #先定位出 1!yyy!mm!dd! 該行, 確認過每個case裡lines[2]都有'1!10'
    
    input_text = lines[2]
    regex = '!\d{4}!'  
    pattern = re.compile(regex)
    num4 = re.findall(pattern, input_text) # 只有一組數字的有144件,沒數字的有26784件, 多組數字的不存在
    if len(num4) == 1: #得到最終判刑月數,共144件
        match_1[case] = num4[0]
        file_names.append( lines[0].split('\\')[-1] )
        jails.append( jail_int(num4[0][1:-1]) ) #拿掉驚嘆號
    else: #再往下探討, (1!下一行)最先出現1.1!的該行取出判刑月數
        for line in lines:
            if line[:4] == '1.1!': #得26751件
                input_text = line
                regex = '有期徒刑!\d{4}!'  
                pattern = re.compile(regex)
                num4_2 = re.findall(pattern, input_text) #無超過1的
                match_2[case] = num4_2
                if num4_2 ==[]: #無罪 不受理 免除其刑,共10件
                    break
                file_names.append( lines[0].split('\\')[-1] )
                jails.append( jail_int(num4_2[0][5:-1]) ) #拿掉有期徒刑!,!
                break

jails[10938] = 4
jails[18004] = 2
#兩個dist ,key為檔名,value為判刑月數, 各建dataframe 再垂直合併
#d = {'file_name':[r1,r2,...],'jail':[r1,r2,...]}
```
